```
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = 'https://raw.githubusercontent.com/mfdmitriy/Python.Matplotlib-Seaborn/main/california_housing_test.csv'
df = pd.read_csv(data, sep = ',', quotechar='"')  # заводим разделитель в файле и обозначение названий
df.head()
```

```
df.info()  # информация о колонках, количестве заполненных данных и типах данных в колонках
```

```
median_income = df.loc[:, 'median_income'].values  # заводим в переменную значения медианного дохода
total_rooms = df.loc[:, 'total_rooms'].values  # заводим в переменную значения количества комнат 
total_bedrooms = df.loc[:, 'total_bedrooms'].values
```

```
plt.plot(median_income, total_rooms)  # зависимость медианного дохода от кол-ва комнат
```

```
plt.style.available  # смотри доступные нам стили
```

```
plt.style.use('seaborn-darkgrid')  # пример использования стиля
plt.plot(median_income, total_rooms)
plt.plot(median_income, total_bedrooms)
```

```
# задаем параметры цвета самостоятельно + размер метки данных
plt.plot(median_income, total_rooms, c = 'c', marker = '.', markersize = 5)
plt.plot(median_income, total_bedrooms, c = 'b', marker = '.', markersize = 5)
```

<b> Boxplot </b>

```
df = pd.read_csv('https://raw.githubusercontent.com/mfdmitriy/Python.Matplotlib-Seaborn/main/wisconsinBreastCancer.csv')
df.head()
```

```
malignant = df.loc[df['diagnosis'] == 'M', 'area_mean'].values  # pandas серия по диагнозу по признаку M, считаем среднее по округу
benign = df.loc[df['diagnosis'] == 'B', 'area_mean'].values

plt.boxplot([malignant, benign], labels = ['M', 'B'])
```

```
df.boxplot(column = 'area_mean', by = 'diagnosis')  # среднее по области по диагнозу
```

```
df.boxplot(column = 'area_mean', by = 'diagnosis')
plt.title('')
plt.suptitle('')
```

<b> Boxplot с помощью Seaborn </b>

```
sns.boxplot(x='diagnosis', y='area_mean', data=df)
```

<b> Scatter Plot </b>

```
plt.scatter(df['radius_mean'], df['area_mean'])
plt.xlabel('radius mean')
plt.ylabel('area mean')
plt.title('area mean by radius.mean')
plt.show()
```

<b> Barplot </b>

```
name = ['A', 'B', 'C', 'D']
age = [10, 20, 50, 100]
df = pd.DataFrame(data={'name' :name, 'age':age})


plt.bar(df['name'], df['age'])
plt.xlabel('name')
plt.ylabel('age')
plt.title('age by name')
plt.show()
```

<b> Круговая диаграмма </b>

```
plt.style.use('seaborn')
plt.pie(df['age'], labels = df['name'],
        autopct = '%1.0f%%')
plt.legend()
plt.title('Age by name', fontsize = 14)
```

<b> Heatmap </b>

```
income = np.random.randint(100, 301, size = (5, 6))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
years = [2017, 2018, 2019, 2020, 2021, 2022]

data = pd.DataFrame(income, index = months, columns = years)  # матрица по рандомным значениям, по строкам месяцы, по столбцам года
data
```

```
sns.heatmap(data)  # тепловая карта, в которой значение отображается с помощью цвета. Самый темный цвет - самый плохой по доходу, самый светлый - самый высокий по доходу
```
